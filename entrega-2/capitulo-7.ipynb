# Capítulo 7: O método do lugar das raízes

O lugar das raízes é um método gráfico utilizado para analisar e projetar sistemas de controle com retroação. Ele consiste em traçar o caminho das raízes da equação característica no plano complexo à medida que um parâmetro do sistema é variado. Esse diagrama, conhecido como diagrama do lugar das raízes, é uma ferramenta poderosa para entender a estabilidade relativa e o desempenho do sistema.

## Introdução
A estabilidade relativa e o desempenho de um sistema em malha fechada estão diretamente ligados à localização das raízes da equação característica no plano complexo. Frequentemente, é necessário ajustar parâmetros para posicionar adequadamente essas raízes. O método do lugar das raízes permite visualizar como as raízes se movem no plano à medida que um parâmetro é variado, fornecendo insights sobre a sensibilidade do sistema a mudanças nesse parâmetro.

Desenvolvido por Evans em 1948, o método do lugar das raízes é amplamente utilizado na engenharia de controle. Ele permite esboçar graficamente o lugar geométrico das raízes, auxiliando no ajuste de parâmetros para alcançar o desempenho desejado. Além disso, pode ser combinado com o critério de Routh-Hurwitz para uma análise mais robusta da estabilidade do sistema.
